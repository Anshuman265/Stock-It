```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def moneycontrol(url = "https://www.moneycontrol.com/news/business/markets/"):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    news_section = soup.find_all('li', class_='clearfix')
    news_text = []
    # Iterate through each news item and extract the headlines
    for x in news_section:
        news_text.append(f'{x.h2.text}. {x.p.text}')
    df = pd.DataFrame(columns= ["text"])
    df["text"] = news_text
    return df       
```

```python
df = moneycontrol()
```
